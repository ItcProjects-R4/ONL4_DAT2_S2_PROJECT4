### Remove Missing Records and Select Final Columns

```python
final_cols = [
    'listing_id',
    'category',
    'listing_type',
    'property_type',
    'completion_status',
    'price_egp',
    'price_period',
    'city',
    'town',
    'district',
    'subdistrict',
    'lat',
    'lon',
    'bedrooms',
    'bathrooms',
    'area_value',
    'furnished',
    'listing_level',
    'is_premium',
    'is_featured',
    'listed_date',
    'images_count',
    'has_view_360',
    'payment_method',
    'agent_id',
    'broker_id',
    'broker_name'
]

df = df.rename(columns={'Ã¯Â»Â¿listing_id': 'listing_id'})
df = df.dropna(subset=['listing_id']).reset_index(drop=True)
df = df[final_cols].copy()

df.shape
```

### Test

```python
df.head()
```

### Fill Missing Values

```python
df['completion_status'] = df['completion_status'].fillna('unknown')
df['district'] = df['district'].fillna('Not Available')
df['subdistrict'] = df['subdistrict'].fillna('Not Available')
df['furnished'] = df['furnished'].fillna('UNKNOWN')
df['payment_method'] = df['payment_method'].fillna('unknown')
```

### **Test**

```python
df.head()
```

### Convert Bedrooms and Bathrooms

```python
for col in ['bedrooms', 'bathrooms']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col].fillna(df.groupby('property_type')[col].transform('median'))
    df[col] = df[col].fillna(df[col].median()).astype(int)
```

### **Test**

```python
df.head()
```

### **Convert Numeric Columns**

```python
for col in ['price_egp', 'area_value', 'images_count', 'agent_id', 'broker_id']:
    df[col] = pd.to_numeric(df[col], errors='coerce').astype(int)
```

### **Convert InvoiceDate / listed_date**

```python
df['listed_date'] = pd.to_datetime(df['listed_date'], utc=True)
```

### **Round Latitude and Longitude**

```python
from decimal import Decimal, ROUND_HALF_UP

def round_half_up_series(s, ndigits=8):
    q = Decimal('1').scaleb(-ndigits)
    return s.apply(
        lambda x: float(Decimal(str(x)).quantize(q, rounding=ROUND_HALF_UP)) if pd.notna(x) else x
    )

df['lat'] = round_half_up_series(df['lat'], 8)
df['lon'] = round_half_up_series(df['lon'], 8)
```
